```python
X_train, X_test, y_train, y_test = train_test_split(df[['age']],df.bought_insurance,train_size=0.8)
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
X_test
```

```python
y_predicted = model.predict(X_test)
```

```python
model.predict_proba(X_test)
```

```python
model.score(X_test,y_test)
```

```python
import math
def sigmoid(x):
  return 1 / (1 + math.exp(-x))
```

```python
def prediction_function(age):
    z = 0.042 * age - 1.53 # 0.04150133 ~ 0.042 and -1.52726963 ~ -1.53
    y = sigmoid(z)
    return y
```

```python
age = 45
prediction_function(age)
```

```python
age = 30
prediction_function(age)
```

```python
plt.scatter(df.age, df.bought_insurance, color='red', marker='+', label='Actual Data')

age_range = np.linspace(df.age.min(), df.age.max(), 100).reshape(-1, 1)
probability = model.predict_proba(age_range)[:, 1]

plt.plot(age_range, probability, color='blue', linewidth=2, label='Logistic Regression Curve')

plt.xlabel("Age")
plt.ylabel("Probability of Buying Insurance")
plt.title("Logistic Regression Conclusion Graph")
plt.legend()

plt.show()
```
